## Question 1

Compare the mean Regular Unleaded Gasoline price among different cities in 2022. Plot the corresponding bar chart.

```python
df = pd.read_csv('data/fueltypesall1991-2023.csv')
#filter to only show Regular Unleaded Gasoline within 2022 
gasoline = df[(df['Fuel Type'] == "Regular Unleaded Gasoline") & (df['Date'].str.contains('2022'))]
#caluate the mean 
mean = gasoline.mean()
#plot a bar graph 
mean.plot.bar()

plt.xlabel("Cities")
plt.ylabel("Mean Price ")
plt.title("Regular Unleaded Gasoline price among different cities in 2022")
plt.show()
```

## Question 2

Plot the histogram of the Regular Unleaded Gasoline price in Ottawa for 2022. Choose the appropriate parameters.

```python
#using the filtered data frame from before, show the price for Ottawa
ottawa = gasoline[['Date', 'Ottawa']]
#plot histogram 
ottawa.hist(column='Ottawa')

plt.xlabel("Price")
plt.ylabel("Frequency ")
plt.title("Regular Unleaded Gasoline price in Ottawa for 2022")
plt.show()

```

## Question 3

Plot the ECDF (normalized) of the Regular Unleaded Gasoline price in Ottawa for 2022.

```python
#Filtering all values for Ottawa
df = gasoline['Ottawa'].values
#calculating ECDF
x = np.sort(df)
y = np.arange (1, len(df) + 1) / len(df)

#Plotting 
plt.plot(x, y, marker='.', linestyle='none')
plt.title("ECDF(normalized) of the R.U.G. price in Ottawa for 2022")
plt.xlabel("Gasoline Price $")
plt.ylabel("ECDF")
plt.show()
```
